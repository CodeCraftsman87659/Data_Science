```python
print('Number of NaN values per column:')
print(df.isnull().sum())
```

```python
df.dropna(subset=['sl_no', 'gender', 'ssc_p', 'ssc_b', 'hsc_p', 'hsc_b', 'hsc_s', 'degree_p', 'degree_t', 'workex', 'etest_p', 'specialisation', 'mba_p', 'status'], inplace=True)

print('Number of NaN values per column after dropping incomplete records:')
print(df.isnull().sum())
```

```python
print("Rows with NaN in 'salary' column:")
display(df[df['salary'].isnull()])

df['salary'].fillna(0, inplace=True)

print('Number of NaN values per column after imputing salary:')
print(df.isnull().sum())
```

### Justification for NaN Imputation:

*   **Dropping Rows (2 records):** For the initial 2 records with missing values across multiple core columns (`sl_no` to `status`), dropping these rows was justified because they represented incomplete student profiles. Imputing a large number of diverse fields could introduce significant bias and uncertainty, especially when the number of such records is very small relative to the dataset size.

*   **Imputing `salary` with 0 (67 records):** The missing values in the `salary` column directly correspond to individuals who have a `status` of 'Not Placed'. Since these individuals did not secure a placement, their salary from this context is legitimately zero. Replacing NaN with 0 in this case accurately reflects their placement outcome without distorting the data with an arbitrary mean or median, which would be inappropriate for non-placed candidates.

## 4)What kind of relation between salary and mba_p

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.scatterplot(x='mba_p', y='salary', data=df[df['status'] == 'Placed'])
plt.title('Relationship between MBA Percentage (mba_p) and Salary for Placed Students')
plt.xlabel('MBA Percentage (mba_p)')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
placed_students = df[df['status'] == 'Placed']

min_salary_by_specialisation = placed_students.groupby('specialisation')['salary'].min().reset_index()

# Find the specialization with the overall minimum salary
min_salary_overall = min_salary_by_specialisation.loc[min_salary_by_specialisation['salary'].idxmin()]

print("Minimum salary by specialization:")
display(min_salary_by_specialisation)

print(f"\nThe specialization getting the minimum salary is '{min_salary_overall['specialisation']}' with a minimum salary of {min_salary_overall['salary']}.")
```

```python
salary_above_500k = df[df['salary'] > 500000].shape[0]

print(f"Number of students getting above 500000 salary: {salary_above_500k}")
```

```python
from scipy import stats

t_statistic, p_value = stats.ttest_rel(df['etest_p'], df['mba_p'])

alpha = 0.05

print(f"Paired t-test results between etest_p and mba_p:")
print(f"t-statistic: {t_statistic:.3f}")
print(f"p-value: {p_value:.3f}")

if p_value < alpha:
    print(f"\nSince the p-value ({p_value:.3f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference between the mean etest_p and mean mba_p scores.")
else:
    print(f"\nSince the p-value ({p_value:.3f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference between the mean etest_p and mean mba_p scores.")
```

```python
from scipy import stats

placed_students_filtered = df[df['status'] == 'Placed']

salary_sci_tech = placed_students_filtered[placed_students_filtered['degree_t'] == 'Sci&Tech']['salary']

salary_mkt_hr = placed_students_filtered[placed_students_filtered['specialisation'] == 'Mkt&HR']['salary']

t_statistic_ind, p_value_ind = stats.ttest_ind(salary_sci_tech, salary_mkt_hr, equal_var=False)
alpha = 0.05

print(f"Independent t-test results between salary of 'Sci&Tech' degree_t and 'Mkt&HR' specialisation:")
print(f"t-statistic: {t_statistic_ind:.3f}")
print(f"p-value: {p_value_ind:.3f}")

if p_value_ind < alpha:
    print(f"\nSince the p-value ({p_value_ind:.3f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in mean salaries between students with 'Sci&Tech' degree and 'Mkt&HR' specialisation.")
else:
    print(f"\nSince the p-value ({p_value_ind:.3f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in mean salaries between students with 'Sci&Tech' degree and 'Mkt&HR' specialisation.")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

salary_reshaped = df['salary'].values.reshape(-1, 1)

df['salary_standardized'] = scaler.fit_transform(salary_reshaped)

print("Original Salary (first 5 values):\n", df['salary'].head())
print("\nStandardized Salary (first 5 values):\n", df['salary_standardized'].head())
print("\nMean of Standardized Salary: ", df['salary_standardized'].mean())
print("Standard Deviation of Standardized Salary: ", df['salary_standardized'].std())
```

```python
from scipy.stats import norm

salary_min = 700000
salary_max = 900000

original_mean = df['salary'].mean()
original_std = df['salary'].std()

standardized_salary_min = (salary_min - original_mean) / original_std
standardized_salary_max = (salary_max - original_mean) / original_std

points = [standardized_salary_min, (standardized_salary_min + standardized_salary_max) / 2, standardized_salary_max]
pdfs = norm.pdf(points)

print(f"Original salary range: {salary_min} to {salary_max}")
print(f"Standardized salary range: {standardized_salary_min:.3f} to {standardized_salary_max:.3f}")

print("\nProbability Density Function (PDF) values at selected points in the standardized range:")
for i, point in enumerate(points):
    print(f"  At standardized value {point:.3f}, PDF = {pdfs[i]:.5f}")

probability_range = norm.cdf(standardized_salary_max) - norm.cdf(standardized_salary_min)
print(f"\nThe probability of salary being within the range {salary_min} to {salary_max} (standardized {standardized_salary_min:.3f} to {standardized_salary_max:.3f}) is: {probability_range:.5f}")
```

```python
from scipy import stats

sci_tech_students = df[df['degree_t'] == 'Sci&Tech']

t_statistic_sci_tech, p_value_sci_tech = stats.ttest_rel(sci_tech_students['etest_p'], sci_tech_students['mba_p'])

alpha = 0.05

print(f"Paired t-test results for 'Sci&Tech' students (etest_p vs mba_p):")
print(f"t-statistic: {t_statistic_sci_tech:.3f}")
print(f"p-value: {p_value_sci_tech:.3f}")

if p_value_sci_tech < alpha:
    print(f"\nSince the p-value ({p_value_sci_tech:.3f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: For 'Sci&Tech' students, there is a statistically significant difference between the mean etest_p and mean mba_p scores.")
else:
    print(f"\nSince the p-value ({p_value_sci_tech:.3f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: For 'Sci&Tech' students, there is no statistically significant difference between the mean etest_p and mean mba_p scores.")
```

```python
numerical_cols = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']


correlation_matrix = df[df['status'] == 'Placed'][numerical_cols + ['salary']].corr()


salary_correlations = correlation_matrix['salary'].sort_values(ascending=False)

print("Correlation of numerical parameters with salary (for Placed students):\n")
display(salary_correlations)

highest_correlated_param = salary_correlations.drop('salary').idxmax()
highest_correlation_value = salary_correlations.drop('salary').max()

print(f"\nThe parameter most highly correlated with salary is '{highest_correlated_param}' with a correlation of {highest_correlation_value:.3f}.")
```
