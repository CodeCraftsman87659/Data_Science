```python
class Univariate:
    @staticmethod
    def quanQual(df):
        quan = df.select_dtypes(include=['number']).columns.tolist()
        qual = df.select_dtypes(include=['object', 'category']).columns.tolist()
        return quan, qual

```

```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
quan,qual=Univariate.quanQual(dataset)
```
